```python
import requests
from bs4 import BeautifulSoup
```

```python
url  = r"https://maktabkhooneh.org/learn/python/?p=1"
```

```python
page = requests.get(url)
if page.status_code == 200:
    print("page founded")
else :
    print("page not found")
```

```python
soup = BeautifulSoup(page.content , "html.parser")
```

```python
links=soup.find_all( class_="course-card__wrapper",href=True)
links
```

```python
link_pages = [link['href'] for link in links if link['href']]

link_pages
```

```python
link_pages=[]

for link in links:
    if link["href"]:
        link_pages.append( link["href"])
        
else:
    for i in link_pages:
        print(i)
```

```python
main_link = "https://maktabkhooneh.org"

test_link = main_link + link_pages[0]

# try:
html = requests.get(test_link)
html_status = html.status_code == 200

course_info = dict()

if html_status:
    soup = BeautifulSoup(html.content , "html.parser")
    
    # try:
    course_info["title"] =soup.select_one(".nuxt-link-exact-active")
    
    course_info["teacher"] =soup.select_one(".course-page__header-opacity-titles--active")
    course_info["subject"] = soup.select_one(".nuxt-link-active+ .course-page__category-tree-link")
    #course_info["total_time"] =soup.select_one(".course-page__clock-svg+ .flex .flex-1").text.split()[0]
    course_info["total_time"] =soup.select_one(".course-page__clock-svg+ .flex .flex-1")
else:
    print("cont connect")
print(course_info)

```

```python
for k,v in course_info.items():
    print(k , v.text)
```
