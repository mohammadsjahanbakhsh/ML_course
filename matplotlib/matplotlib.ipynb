```python
# x axis values 
x = [1,2,3,4,5,6] 
# corresponding y axis values 
y = [2,4,1,5,2,6] 
# plotting the points 
plt.plot(x, y, color='green', linestyle='dotted', linewidth = 3 ,marker = 'o', markersize = 8 , markerfacecolor = "red") 
# setting x and y axis range 
plt.ylim(1,8) 
plt.xlim(1,8) 
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('Some cool customizations!') ;
```

```python
x = np.array([0,1,2,3,4])
y = np.array([0,2,4,6,8])

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
plt.plot(x,y, label='2x', color='blue', linewidth=2, marker='.', linestyle='--', markersize=10, markerfacecolor = "red")


#plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


plt.xlim(1,5) 


# Add a legend
plt.legend()

# Save figure (dpi 200 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=200) #  DPI stands for dots per inch
```

```python
# x-coordinates of left sides of bars   
Labels = ['one', 'two', 'three', 'four', 'five']  
# heights of bars   
height = [10, 20, 30, 40, 50]   
 
# plotting a bar chart   
plt.bar(Labels,height, color = ['blue', 'green', 'red','yellow','black'],width = 0.6 , alpha = 0.8)

# naming the x-axis   
plt.xlabel('x - axis')   
# naming the y-axis   
plt.ylabel('y - axis')   
# plot title   
plt.title('bar chart!');
```

```python
# defining labels 
activities = ['eat', 'sleep', 'work', 'play']

# portion covered by each label 
slices = [3, 7, 8, 6] 

# color for each label 
colors = ['r', 'y', 'g', 'b'] 

# plotting the pie chart 
plt.pie(slices, labels = activities, colors=colors, explode = (0, 0, 0, 0.1), radius = 1.2 , autopct = '%.2f%%') # explode a pie chart means to make one of the wedges of the pie chart to stand out.

# plotting legend 
plt.legend(loc = 1) ;
```

```python
fig, ax = plt.subplots()

x = np.array([1,2,3,4,5])
y = x * 2

ax.plot(x + 2 ,y,marker = 'x')                      # blue
ax.plot(x + 1 ,y,marker = 'o')                      # orange
ax.plot(x ,y ,marker = 'p')                         # green
ax.plot(x - 1 ,y,marker = '*', markersize = 5)      # red
ax.plot(x - 2 ,y,marker = '1', markersize = 10)     # purple
ax.plot(x - 3 ,y,marker = 'v', markersize = 15)     # brown
ax.plot(x - 4 ,y,marker = 'h', markersize = 25);    # pink
```
