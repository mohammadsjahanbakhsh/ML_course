```python
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import classification_report
```

```python
param_grid = {
    'max_depth': [5, 15, 25],
    'min_samples_leaf': [1, 3],
    'max_leaf_nodes': [10, 20, 35, 50]}
```

```python
data.data
```

```python
X = data.data
y = data.target
```

```python
dt = DecisionTreeClassifier()
gs = GridSearchCV(dt, param_grid,
                   scoring='f1', cv=5)
gs.fit(X, y);
```

```python
print("best params:", gs.best_params_)
print("best score:", gs.best_score_)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import precision_score,recall_score,f1_score,confusion_matrix
import numpy as np
from sklearn.model_selection import train_test_split

kf = KFold(n_splits=5, shuffle=True, random_state=313)
dt_accuracy_scores = []
dt_precision_scores = []
dt_recall_scores = []
dt_f1_score=[]

for train_index, test_index in kf.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    y_train, y_test = y[train_index], y[test_index]
    dt = DecisionTreeClassifier(max_depth=5, max_leaf_nodes=50, min_samples_leaf= 1)
    dt.fit(X_train, y_train)
    dt_accuracy_scores.append(dt.score(X_test, y_test))
    dt_y_pred = dt.predict(X_test)
    dt_precision_scores.append(precision_score(y_test, dt_y_pred))
    dt_recall_scores.append(recall_score(y_test, dt_y_pred))
    dt_f1_score.append(f1_score(y_test, dt_y_pred))

print("Decision Tree")
print("  accuracy:", np.mean(dt_accuracy_scores))
print("  precision:", np.mean(dt_precision_scores))
print("  recall:", np.mean(dt_recall_scores))
print("  f1_score:", np.mean(dt_f1_score))
```
